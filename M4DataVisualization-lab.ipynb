# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")
df = pd.read_sql_query("SELECT * FROM master", conn)

# Drop missing values from the 'ConvertedComp' column
converted_comp = df['ConvertedComp'].dropna()

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(converted_comp, bins=30, color='blue', alpha=0.7)
plt.title('Histogram of Converted Compensation (USD)')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Load the data from the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")
df = pd.read_sql_query("SELECT * FROM master", conn)

# Drop missing values from the 'Age' column
age_data = df['Age'].dropna()

# Plot box plot
plt.figure(figsize=(10, 6))
plt.boxplot(age_data, vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Load the data from the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")
df = pd.read_sql_query("SELECT * FROM master", conn)

# Drop missing values from the 'Age' and 'WorkWeekHrs' columns
age_workweek_data = df[['Age', 'WorkWeekHrs']].dropna()

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(age_workweek_data['Age'], age_workweek_data['WorkWeekHrs'], alpha=0.5, color='purple')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Load the data from the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")
df = pd.read_sql_query("SELECT * FROM master", conn)

# Drop missing values from the relevant columns
bubble_data = df[['WorkWeekHrs', 'CodeRevHrs', 'Age']].dropna()

# Plot bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(
    bubble_data['WorkWeekHrs'],
    bubble_data['CodeRevHrs'],
    s=bubble_data['Age'] * 10,  # Adjusting the bubble size for better visualization
    alpha=0.5,
    color='orange'
)
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size = Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load the data from the 'DatabaseDesireNextYear' table
df_database_desire = pd.read_sql_query("SELECT * FROM DatabaseDesireNextYear", conn)

# Display the first few rows of the DataFrame
df_database_desire.head()
```

```python
# your code goes here
# Calculate the count of each database respondents wish to learn
database_counts = df_database_desire['DatabaseDesireNextYear'].value_counts().head(5)

# Plot pie chart
plt.figure(figsize=(10, 6))
plt.pie(database_counts, labels=database_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")
df = pd.read_sql_query("SELECT * FROM master", conn)

# Filter data for the age group 30 to 35
age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median values
median_work_week_hrs = age_group['WorkWeekHrs'].median()
median_code_rev_hrs = age_group['CodeRevHrs'].median()

# Create a DataFrame for the median values
median_data = pd.DataFrame({
    'Category': ['WorkWeekHrs', 'CodeRevHrs'],
    'Median Hours': [median_work_week_hrs, median_code_rev_hrs]
})

# Plot stacked bar chart
plt.figure(figsize=(10, 6))
median_data.set_index('Category').T.plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age Group 30 to 35')
plt.ylabel('Median Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")
df = pd.read_sql_query("SELECT * FROM master", conn)

# Filter data for the age group 45 to 60
age_filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_converted_comp = age_filtered_df.groupby('Age')['ConvertedComp'].median().dropna()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o', linestyle='-', color='blue')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp (USD)')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")
df = pd.read_sql_query("SELECT * FROM master", conn)

# Calculate the count of each main branch
main_branch_counts = df['MainBranch'].value_counts()

# Plot horizontal bar chart
plt.figure(figsize=(12, 8))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of Respondents by Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```

## In the list of most popular languages respondents wish to learn next year, what is the rank of Python?

```python
import pandas as pd
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load the data from the 'LanguageDesireNextYear' table
df_language_desire = pd.read_sql_query("SELECT * FROM LanguageDesireNextYear", conn)

# Calculate the count of each language respondents wish to learn
language_counts = df_language_desire['LanguageDesireNextYear'].value_counts()

# Convert the counts to a DataFrame for easy manipulation
language_counts_df = language_counts.reset_index()
language_counts_df.columns = ['Language', 'Count']

# Determine the rank of Python
language_counts_df['Rank'] = language_counts_df['Count'].rank(ascending=False, method='min')
python_rank = language_counts_df.loc[language_counts_df['Language'] == 'Python', 'Rank'].values[0]

# Print the rank of Python
print("The rank of Python in the list of most popular languages respondents wish to learn next year is:", int(python_rank))
```

## Question 3How many respondents indicated that they currently work with 'SQL'? 

```python
import pandas as pd
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load the data from the 'LanguageWorkedWith' table
df_language_worked = pd.read_sql_query("SELECT * FROM LanguageWorkedWith", conn)

# Filter the rows where respondents indicated they work with SQL
sql_respondents = df_language_worked[df_language_worked['LanguageWorkedWith'].str.contains('SQL', na=False)]

# Count the number of respondents who work with SQL
sql_count = sql_respondents.shape[0]

# Print the number of respondents
print("Number of respondents who indicated that they currently work with 'SQL':", sql_count)
```

## How many respondents indicated that they work on 'MySQL' only? 

```python
import pandas as pd
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Load the data from the 'LanguageWorkedWith' table
df_language_worked = pd.read_sql_query("SELECT * FROM LanguageWorkedWith", conn)

# Filter the rows where respondents indicated they work with 'MySQL'
mysql_respondents = df_language_worked[df_language_worked['LanguageWorkedWith'].str.contains('MySQL', na=False)]

# Further filter rows to check if they work with 'MySQL' only
mysql_only_respondents = mysql_respondents[mysql_respondents['LanguageWorkedWith'] == 'MySQL']

# Count the number of respondents who work with 'MySQL' only
mysql_only_count = mysql_only_respondents.shape[0]

# Print the number of respondents
print("Number of respondents who indicated that they work on 'MySQL' only:", mysql_only_count)
```
